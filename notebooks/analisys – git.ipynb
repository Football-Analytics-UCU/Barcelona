```python
# database match
match = pd.read_sql(
    """SELECT * FROM Match;""", conn)

match
```

* 1. Аналіз перемог, якщо матч "вдома" або "на виїзді"

* 2. Залежність перемоги від воротарів
- аналіз кількості перемог коли воротар Хосе Мануэль Пінто(2008-2014)
- аналіз кількості перемог коли воротар Рубен Міньо(2010-2012)
- аналіз кількості перемог коли воротар Клаудіо Браво(2014-2016)

* АБО купили в якомусь році "дорогого" ігрока, як змінилася кількість перемог в роки після цього(корисний чи ні) - знатоки футбола потрібна допомога

* АБО зміна тренера в якомусь році - аналіз перемог(перемога, поразка, нічия) за кожного наступного тренера по рокам(самий ефективний тренер)
   
* 3. Залежність перемог від суперника(наприклад в грі з Granada CF - 70% поразка, з CD Numancia - 20%)

* 4. Зміна поля для тренування(вплив на фіз підготовку) в якомусь році - з ....року почали більше програвати

```python
leagues = pd.read_sql("""SELECT *
                        FROM League
                        JOIN Country ON Country.id = League.country_id;""", conn)
leagues
```

```python
# matches FC Barcelona
detailed_matches = pd.read_sql("""SELECT Match.id, 
                                        League.name AS league_name,                                         
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name AS  home_team,
                                        AT.team_long_name AS away_team,
                                        home_team_goal, 
                                        away_team_goal                                        
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT ON HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT ON AT.team_api_id = Match.away_team_api_id
                                WHERE (HT.team_long_name = 'FC Barcelona' OR AT.team_long_name = 'FC Barcelona')
                                ORDER by date;""", conn)
detailed_matches
```
